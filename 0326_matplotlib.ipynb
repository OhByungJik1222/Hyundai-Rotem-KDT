```python
# 제공된 excel 파일을 이용해서
# DataFrame을 생성하고 적절한 line plot을 그려보아요!
# excel 파일을 불러들여서 처리해야해요!
# excel 파일을 로딩하기 위해서는 openpyxl 모듈이 필요해요!
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('C:\education\현대로템\lineplot_sample_data.xlsx')
display(df.head(20))
print()

# DataFrame을 가져와 봤더니 전출지별 컬럼에 NaN(결측치)가 포함되어 있어요!
# 변환과정에서 약간 문제가 있어서 값이 NaN으로 설정되었는데 바로 앞의 행의 값으로
# NaN값을 변경시켜 주면 될 것 같아요!
# df = df.fillna(method='ffill')
df = df.ffill() # forward fill
display(df.head(30))
```

```python
# 우리가 관심있는 정보는
# 서울특별시에서 다른 지역으로 이동한 데이터이터.
# 전출지별 컬럼에서 서울특별시 라는 값을 갖는 데이터만 추출
# 전입지별 컬럼에서는 서울특별시 값을 제외시킬거에요! (서울에서 다른 도시로 이동한 사람) 서울 -> 서울 X
seoul_df = df.loc[(df['전출지별'] == '서울특별시') & (df['전입지별'] != '서울특별시')]
display(seoul_df)
```

```python
# 추가적으로 몇가지 작업만 더 해봐요!
# '전출지별' 컬럼은 필요가 없으니까 삭제할거구요!
# '전입지별' 컬럼의 이름은 '전입지'로 변경할거에요!
# 기본적으로 숫자 인젝스로 되어있는데 '전입지' 컬럼을 지정 index로 설정할거에요!
seoul_df = seoul_df.drop('전출지별', axis=1, inplace=False)
seoul_df = seoul_df.rename({'전입지별':'전입지'}, axis=1, inplace=False)
seoul_df = seoul_df.set_index('전입지', inplace=False)
display(seoul_df)
```

```python
# 이제 선그래프를 그리기 위해서 경기도만 추출할거에요!
# (서울에서 경기도로 전입한 사람의 연도별 추이)
s = seoul_df.loc['경기도']
print(s)
```

```python
# 서울에서 경기도로 전입한 사람들의 연도별 수를 알아냈어요!(Series)
# 이 데이터를 이용해서 line plot을 그려보아요!
plt.plot(s.index, s.values)
plt.show()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib import rc
import numpy as np
import pandas as pd

mpl.rcParams['axes.unicode_minus'] = False

font_path = 'C:\education\현대로템\malgun.ttf'
font_name = fm.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 그래프의 크기를 지정할 수 있어요!
plt.figure(figsize=(10,5)) # 그림의 가로길이 세로길이를 tuple로 명시(단위는 inch)

plt.title('서울 -> 경기 인구 이동')
plt.ylabel('이동 인구 수')
plt.xlabel('연도')

# 그래프의 스타일을 지정할 수도 있어요!
plt.style.use('ggplot')

# x축 눈금(연도)가 겹쳐서 알아보기 어려워요!
# label을 회전시켜서 겹쳐보이지 않게 처리
plt.xticks(rotation='vertical')

# 그래프를 그릴때 약간의 속성들을 줄 수 있어요!
# marker나 색상, 추가적인 문자열을 사용할 수 있어요!
plt.plot(s.index, s.values, marker='o', markersize=5, markerfacecolor='y', color='b', linewidth=3)

# 범례
plt.legend(labels=['서울 -> 경기'], loc='best')

# y축 범위를 증가시켰어요. 주석을 포함할 공간을 확보할 목적으로..
plt.ylim(5000, 800000)

# 주석과 화살표를 표시
plt.annotate("", # 화살표 시작부분에 문자열을 추가할 수 있어요! 일반적으로 안씀
            xy=(20, 620000), # 화살표의 끝점. x좌표는 index, y좌표는 값을 이용
            xytext=(2, 290000), # 화살표의 시작점.
            xycoords='data', # 좌표의 기준
            arrowprops=dict(arrowstyle='->', color='skyblue', lw=5)) # lw (line width)

# 텍스트를 표시
plt.annotate("인구 이동 증가(1970-1995)",
            xy=(10, 350000), # 텍스트 위치 기준점
            rotation=30, # 텍스트 회전
            va='baseline',
            ha='center',
            fontsize=15)

plt.annotate("",
            xy=(47, 450000),
            xytext=(30, 620000),
            xycoords='data',
            arrowprops=dict(arrowstyle='->', color='olive', lw=5))

plt.annotate("인구 이동 감소(1995-2017)",
            xy=(39, 500000),
            rotation=-18,
            va='baseline',
            ha='center',
            fontsize=15)

plt.show()
```

```python
# 여기까지 해서 line plot의 여러가지 기능에 대해서 알아보았어요!
# 그 다음 살펴볼 내용은 figure와 subplot이에요!
# 도화지(canvas -> figure)에 여러개의 plot을 그릴 수 있어요!
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
from matplotlib import rc
import numpy as np
import pandas as pd

mpl.rcParams['axes.unicode_minus'] = False

font_path = 'C:\education\현대로템\malgun.ttf'
font_name = fm.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 그래프의 스타일을 지정할 수도 있어요!
plt.style.use('ggplot')

# 그래프의 크기를 지정할 수 있어요!
fig = plt.figure(figsize=(10,10)) # 그림의 가로길이 세로길이를 tuple로 명시(단위는 inch)
# figure에 그래프를 그릴 수 있는 영역을 추가해요!
ax1 = fig.add_subplot(2, 1, 1) # 2행 1열 1번째
ax2 = fig.add_subplot(2, 1, 2) # 2행 1열 2번째

ax1.set_title('서울 -> 경기 인구 이동', size=20)
ax1.set_xlabel('기간', size=12)
ax1.set_ylabel('이동 인구수', size=12)

ax1.plot(s.index, s.values, 'o', markersize=5)

ax2.plot(s.index, s.values, marker='o', markersize=5, markerfacecolor='b')

plt.show()
```

```python
# Box plot에 대해서 알아보아요!
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/KDT/data/auto-mpg.csv', header=None)
df.columns = ['MPG', 'Cylinder', 'Displacement', 'Horsepower', 'Weight', 'Acceleration', 'Year', 'Origin', 'Name']

plt.style.use('ggplot')

# 미국 생산 자동차, 유럽 생산 자동차, 일본 생산 자동차를 가지고
# Box plot을 세개 그릴거에요!
# MPG값을 이용해서 boxplot을 그릴거애요!
plt.boxplot(x=[df.loc[df['Origin'] == 1,'MPG'],
               df.loc[df['Origin'] == 2,'MPG'],
               df.loc[df['Origin'] == 3,'MPG']],
            tick_labels=['USA', 'EU', 'JPN'])

plt.show()
```
