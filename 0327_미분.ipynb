```python
# 함수를 하나 만들거에요!
# 미분하는 함수를 하나 만들거에요!
# 당연히 이 함수는 인자로 함수로 받아야해요!
# def numerical_derivative(미분할 함수, 미분을 구할 점)
def numerical_derivative(f, x):
    # 중앙차분을 이용한 미분값을 계산
    delta_x = 0.0001

    return (f(x + delta_x) - f(x - delta_x)) / (2 * delta_x)

# f(x) = x^2
def my_func(x):
    return x**2

result = numerical_derivative(my_func, 3)
print(result)
```

```python
# 좋은 질문입니다! 함수의 변화량에 따라 delta_x 값이 달라지는 이유는 미분의 정의와 수치 미분의 한계 때문입니다.

# 1. 미분의 정의:

# 미분은 함수의 특정 지점에서의 순간 변화율을 나타냅니다. 이 순간 변화율을 구하기 위해서는 함수의 두 점 사이의 기울기를 계산하고, 두 점 사이의 거리를 0에 가깝게 보내는 극한을 취해야 합니다.

# 수식으로 표현하면 다음과 같습니다.


# f'(x) = lim (h->0) (f(x + h) - f(x)) / h
# Use code with caution
# 여기서 h는 두 점 사이의 거리를 나타내며, delta_x와 같은 의미입니다.

# 2. 수치 미분의 한계:

# 컴퓨터는 극한을 직접 계산할 수 없기 때문에, 수치 미분에서는 delta_x를 매우 작은 값으로 설정하여 미분을 근사합니다. 하지만 delta_x를 너무 작게 설정하면 반올림 오차가 발생하여 오히려 미분값이 부정확해질 수 있습니다.

# 3. 함수의 변화량과 delta_x:

# 함수의 변화량이 클수록, 즉 함수의 기울기가 급격하게 변할수록 delta_x를 작게 설정해야 미분값을 정확하게 근사할 수 있습니다.

# 변화량이 큰 경우: delta_x가 크면 두 점 사이의 기울기가 실제 순간 변화율과 크게 차이가 날 수 있습니다. 따라서 delta_x를 작게 설정하여 두 점을 더 가깝게 만들어야 합니다.
# 변화량이 작은 경우: delta_x가 크더라도 두 점 사이의 기울기가 실제 순간 변화율과 크게 차이가 나지 않습니다. 따라서 delta_x를 크게 설정해도 비교적 정확한 미분값을 얻을 수 있습니다.
# 요약:

# 함수의 변화량이 클수록 delta_x를 작게 설정해야 미분값을 정확하게 근사할 수 있습니다.
# delta_x를 너무 작게 설정하면 반올림 오차가 발생하여 미분값이 부정확해질 수 있습니다.
# 따라서 함수의 변화량에 따라 delta_x 값을 적절하게 조정해야 합니다.

# 함수의 기울기에 따른 delta_x 설정
import numpy as np

# 변화가 심한 함수 예시: f(x) = sin(1/x)
def f(x):
    return np.sin(1/x)

# 작은 delta_x 값 (예: 0.00001) 사용
delta_x = 0.00001

# x = 0.1 에서의 미분값 계산
x = 0.1
derivative = (f(x + delta_x) - f(x - delta_x)) / (2 * delta_x)

print(f"x = {x} 에서의 미분값: {derivative}")

# ------------------------------------------------------------------------------

import numpy as np

# 변화가 완만한 함수 예시: f(x) = x^2
def f(x):
    return x**2

# 큰 delta_x 값 (예: 0.01) 사용
delta_x = 0.01

# x = 2 에서의 미분값 계산
x = 2
derivative = (f(x + delta_x) - f(x - delta_x)) / (2 * delta_x)

print(f"x = {x} 에서의 미분값: {derivative}")

# ------------------------------------------------------------------------------

import autograd.numpy as np
from autograd import grad

# 자동 미분을 위한 함수 정의
def f(x):
    return np.sin(1/x)

# 자동 미분 함수 생성
gradient = grad(f)

# x = 0.1 에서의 미분값 계산
x = 0.1
derivative = gradient(x)

print(f"x = {x} 에서의 미분값: {derivative}")
```

```python
# 다변수를 가지고 있는 함수의 편미분
import numpy as np

def numerical_derivative(f, x):
    # f : 미분을 하려는 함수 (다변수 함수)
    # x : 모든 변수를 포함하고 있는 ndarray(차원 무관) ex) [1.0, 2.0]

    delta_x = 0.0001
    derivative_x = np.zeros(x.shape) # [0.0, 0.0]

    it = np.nditer(x, flags=['multi_index']) # iterator를 하나 생성해요!

    while not it.finished: # 첫번째 변수에대한 예시임!
        idx = it.multi_index # 1.0

        tmp = x[idx] # 첫번째 index에 대한 값을 임시로 저장 (1.0)

        x[idx] = tmp + delta_x # [1.0001, 2.0]
        fx_plus_delta = f(x) # f(x + delta_x)

        x[idx] = tmp - delta_x # [0.9999, 2.0]
        fx_minus_delta = f(x) # f(x - delta_x)

        derivative_x[idx] = (fx_plus_delta - fx_minus_delta) / (2 * delta_x)

        x[idx] = tmp # 원복

        it.iternext()

    return derivative_x

# 일단 1변수 함수부터 미분을 수행해 보아요!
def my_func1(x):
    return x**2

result1 = numerical_derivative(my_func1, np.array([3.0]))
print(result1) # [6.]

# 이번에는 2변수 함수 미분을 수행해 보아요!
def my_func2(input_object):
    x = input_object[0]
    y = input_object[1]
    return 2*x + 3*x*y + np.power(y, 3) # 2x + 3xy + y^3

result2 = numerical_derivative(my_func2, np.array([1.0, 2.0]))
print(result2)

# 3변수 함수
def my_func3(input_object):
    x = input_object[0]
    y = input_object[1]
    z = input_object[2]
    return 2*x*z + 3*x*np.power(y, 2)*z + 7*np.power(z, 3)

result3 = numerical_derivative(my_func3, np.array([[1.0], [2.0], [3.0]]))
print(result3)
```
