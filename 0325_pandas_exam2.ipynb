```python
# 2.
df_seoul['일교차'] = df_seoul['최고기온(℃)'] - df_seoul['최저기온(℃)']
print(df_seoul.loc[df_seoul['일교차'] == df_seoul['일교차'].max()]['날짜'].values[0])
print(df_seoul.loc[df_seoul['일교차'] == df_seoul['일교차'].max()]['일교차'].values[0])
```

```python
# 3.
df_seoul['날짜'] = pd.to_datetime(df_seoul['날짜'])
df_seoul['월일'] = df_seoul['날짜'].dt.month.astype(str) + '월 ' + df_seoul['날짜'].dt.day.astype(str) + '일'

df_seoul_md = df_seoul.groupby('월일')
print(df_seoul_md.일교차.mean().idxmax())
print(format(df_seoul_md.일교차.mean().max(), '.2f'))
```

```python
df_seoul = pd.read_csv('/content/drive/MyDrive/KDT/data/seoul.csv', skiprows=7)
df_daegu = pd.read_csv('/content/drive/MyDrive/KDT/data/daegu.csv', skiprows=7)

df_seoul = df_seoul.dropna(axis=0, how='any')
df_daegu = df_daegu.dropna(axis=0, how='any')

df = pd.merge(df_seoul, df_daegu, on='날짜')
df['서울_대구'] = df['평균기온(℃)_x'] > df['평균기온(℃)_y']
display(df)

df['날짜'] = pd.to_datetime(df['날짜'])
df['연도'] = df['날짜'].dt.year
new_df = df.loc[:, ['연도', '서울_대구']]
display(new_df.head())
```

```python
# 4.
new_df_grouped = new_df.loc[new_df['서울_대구']].groupby('연도')
print(new_df_grouped.count().idxmax().values[0])
print(new_df_grouped.count().max().values[0])
```
