```python

# mdarray의 생성 및 타입 확인
# 일단 설치되어 있는지 확인해 보아요!!
import numpy as np

print(f'numpy의 버전 : {np.__version__}')
```

```python

# numpy가 설치되어 있으니 이제 ndarray를 생성해 보아요!

# list를 literal로 표현하면 [].
a = [1, 2, 3, 4]
print(f'list : {a}')
print(f'list의 type : {type(a)}')
print()

# ndarray를 만들어 보아요! ndarray는 literal이 없어요!
# 배열이기 때문에 모든 요소의 데이터 타입이 같아요!
arr = np.array([1, 2, 3, 4])
print(f'ndarray : {arr}')
print(f'ndarray의 type : {type(arr)}')
print(f'ndarray의 dtype : {arr.dtype}')
print()

# 하나의 ndarray에는 하나의 data type만 들어갈 수 있기 때문에 모두 문자열로 변환됨.
# <U32 -> 32비트 유니코드
arr = np.array([100, 3.14, "Hello", True])
print(f'ndarray : {arr}')
print(f'ndarray의 type : {type(arr)}')
print(f'ndarray의 dtype : {arr.dtype}')
print(f'ndarray의 첫번째 요소의 type : {type(arr[0])}')
```

```python

# 다차원 ndarray 생성
my_list = [[1, 2, 3], [4, 5, 6]]
arr = np.array(my_list)
print(arr)
print()

# 현재 2차원 형태의 ndarray를 만들었어요!
# indexing을 이용해 특정 요소를 출력!
# 1행 2열의 값을 출력해 보아요!
print(f'1행 2열의 값은 : {arr[1, 2]}')
print(f'1행의 값은 : {arr[1]}')
print(f'1행 2열의 값은 : {arr[1][2]}')
```

```python

# ndarray 생성 시 dtype을 설정할 수 있어요!
arr = np.array([1, 2, 3, 4])
print(f'ndarray의 dtype : {arr.dtype}') # int64
print()

arr = np.array([1, 2, 3, 4],
               dtype=np.float64)
print(arr)
print(f'ndarray의 dtype : {arr.dtype}') # float64
```

```python
# list를 이용해서 ndarray를 생성하는 방법에 대해서 알아보았어요!
```

```python

# ndarray가 가지고 있는 속성에 대해 알아보아요!
# ndim, shape이라는 속성 두개만 일단 기억하면 될 것 같아요!

# shape은 차원의 개수와 각 차원의 요소를 표현해 놓은 것.
# shape은 무조건 python의 tuple로 표현.
# 예를 들어 1차원이면 tuple의 요소가 1개
#           2차원이면 tuple의 요소가 2개
#           3차원이면 tuple의 요소가 3개
my_list = [1, 2, 3, 4]
arr = np.array(my_list)
print(f'arr이 가지고 있는 ndim의 값 : {arr.ndim}')
print(f'arr이 가지고 있는 shape의 값 : {arr.shape}')
print()


# 2차원 ndarray를 만들어서 확인해 보아요!
my_list = [[1, 2, 3], [4, 5, 6]]
arr = np.array(my_list)
print(f'arr이 가지고 있는 ndim의 값 : {arr.ndim}')
print(f'arr이 가지고 있는 shape의 값 : {arr.shape}')
```

```python
# 실습1
# 모든 원소의 값이 1인 shape이 (3, 4, 2, 3)인 ndarray를 생성해보세요!

my_list = [[[[1 for _ in range(3)] for _ in range(2)] for _ in range(4)] for _ in range(3)]
arr = np.array(my_list)
print(arr.shape)
```

```python
# 실습 2
# shape이 (2, 2, 3)이고 모든 요소의 값이 0인 ndarray를 생성해보세요!
# 단, list를 이용해서 생성해보세요!

my_list = [[[0 for _ in range(3)] for _ in range(2)] for _ in range(2)]
arr = np.array(my_list)
print(arr.shape)
```

```python

# ndim과 shape에 대해서 알아보았어요!
# size 속성도 있어요. => 총 요소의 개수

my_list = [[[0 for _ in range(3)] for _ in range(2)] for _ in range(2)]
arr = np.array(my_list)
print(arr.shape)
print(arr.size)
```

```python

#  한가지 주의해야 하는게 있어요!
# python의 기본 함수 중 len()이 있어요!
# python의 함수이기 때문에 list에 대해서 len()을 적용하면
# list의 길이를 알려줘요! (중첩 리스트가 있건 없건 상관하지 않아요!)

my_list = [[[0 for _ in range(3)] for _ in range(2)] for _ in range(2)]
arr = np.array(my_list)
print(arr.shape)
print(arr.size)
print(len(arr))
```

```python
# shape 속성은 read only가 아니에요!
# shape을 변경시키면 ndarray의 형태가 바뀌게 되요!

my_list = [1, 2, 3, 4]
arr = np.array(my_list)
print(arr.shape)
print()

arr.shape = (2, 2)
print(arr)
print(arr.shape)
```

```python

#  ndarray의 data type(dtype)을 변환시키는 함수에 대해 알아보아요!

# 실수로 구성되어 있는 1차원 ndarray를 생성해보아요!
my_list = [1.1, 3.14, 10.6, 8.7, 5]
arr = np.array(my_list)
print(arr)
print(arr.dtype)
print()

arr = arr.astype(np.int64)
print(arr)
print(arr.dtype)
```

```python
# 지금까지는 list를 이용해서 ndarray를 생성했어요.
# 당연히 ndarray를 생성하는 다른 방법도 존재해요!
# 하나씩 알아보아요!

# 기본적으로 dtype의 default는 float64로 설정되요.
arr = np.zeros((3, 4), dtype=np.int64)
print(arr)
print()

arr = np.ones((3, 4))
print(arr)
print()

arr = np.empty((3, 5))
print(arr)
print()

# 수업때 다루진 않았지만 데이터 라벨링 작업 시 유용할 것 같음.
arr = np.full((2, 3), 7.0)
print(arr)
```

```python

# python에서 sequence data type =>
# list(실체가 있어요!), tuple, range(실체가 없어요! (메모리 안에 실제 데이터 X))

# numpy에는 arange()가 있어요. python의 range와 비슷.
# numpy의 arange는 실체가 있어요!

# np.arange(시작, 끝, 증감)
#           시작은 inclusive, 끝은 exclusive로 사용.
arr = np.arange(0, 10, 1)
print(arr)
print()

arr = np.arange(20)
print(arr)
```

```python

# numpy가 제공하는 함수를 이용해서 ndarray를 생성하는 방법에 대해 알아보고 있어요!
# ones, zeros, empty, full, arange에 대해 알아보았어요!
# 추가적으로 linear space에 대해서 알아보아요!
# linspace()
# 주어진 범위에서 선형적으로 균등하게 분포된 숫자 배열을 만들어내는데 사용하는데 그 숫자 개수를 이용해서 추출.
import numpy as np
import matplotlib.pyplot as plt

# np.linspace(start, stop, num) <=> np.arange(start, stop, ++)
# 원소간의 간격은 (stop - start) / (num - 1) 으로 계산되요!

arr = np.linspace(0, 10, 11)
print(arr)
print()

arr = np.linspace(0, 10, 21)
print(arr)
plt.plot(arr)
plt.show()
```

```python
# numpy가 가지는 ndarray 생성함수에 대해서 알아보았어요!
# zeros, ones, empty, full, ararnge, linspace
```

```python

# random에 기반한 ndarray 생성 함수가 있어요!
# 총 6개가 있는데 그 중 많이 사용되는 3개만 알아보아요!
# 분포를 기준으로 랜덤값을 추출해서 ndarray로 생성해요!
# (정규분포 혹은 균등분포를 이용)
# 정규분포인 경우는 평균과 표준편차를 정해줘야 해요!
mean = 50
std = 2
arr = np.random.normal(mean, std, (100000,))
print(arr)

# 이렇게 생성된 데이터를 이용해서 histogram을 그려주면
# 데이터가 어떻게 구성되어 있는지 확인할 수 있어요!
# bins는 arr의 min과 max를 bins의 값만큼 영역을 나누는 것.
plt.hist(arr, bins=100)
plt.show()
```

```python

# 표준 정규분포에서 난수를 추출하는 함수에요!
# 평균이 0이고 표준편차가 1인 정규분포
arr = np.random.randn(100000)
print(arr)
plt.hist(arr, bins=1000)
plt.show()
```

```python

# 다른 함수는 모두 실수를 도출하는데
# 정수횽 난수를 추출하는건 딱 1개 있어요!
# 균등분포에서 추출해요!

arr = np.random.randint(-100, 100, (100000,))
print(arr)
plt.hist(arr, bins=100)
plt.show()
```

```python
# shape 변경에 대해서 알아보아요!
# shape을 임의로 변경할 수 있어요!
arr = np.arange(0, 12, 1)
print(arr.shape)
print(arr)
print()

# shape을 직접 변경
# 원본데이터가 실제적으로 변경되요!
# 이런 현상을 매우 싫어해요!
arr.shape = (3, 4)
print(arr.shape)
print(arr)
```

```python
# 그러면 어떻게 하나요?
# shape을 직접적으로 변경하는게 아니라 함수를 이용해서 변경해요!
arr = np.arange(0, 12, 1)
print(arr.shape)
print(arr)
print()

arr1 = arr.reshape(3, 4)
arr2 = arr.reshape(3, 4).copy()
print(arr1)
print(arr2)
print()

# reshape을 이용하면 새로운 ndarray가 생성되는게 아니라 view가 만들어져요!
# arr1은 arr을 (n, m)으로 바꾸어 보여주는 창을 만들어주는 것(값 저장 X)
# copy() 사용 시 값까지 복사 But, 메모리 사용량 증가
arr[0] = 100
print(arr)
print(arr1)
print(arr2)
```
