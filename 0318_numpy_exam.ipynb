```python
import numpy as np
```

```python
## ndarray 생성과 배열 조작
# 1.
arr = np.arange(16)
print(arr)
```

```python
# 2.
arr = np.ones((5, 5))
print(arr)
```

```python
# 3.
arr = np.linspace(0, 1, 20)
print(arr)
```

```python
# 4.
arr = np.random.randn(12).reshape(3, 4).copy()
print(arr)
```

```python
# 5.
arr = np.arange(1, 10).reshape(3, 3).copy()
print(arr)
```

```python
# 6.
arr = np.zeros((5, 5))
arr[1:4,1:4] = 1
print(arr)
```

```python
# 7.
arr = np.arange(1, 10).reshape(3, 3).copy()
print(arr * arr)
```

```python
# 8.
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print(np.vstack((a, b)))
```

```python
# 9.
arr = np.zeros((3, 3))
print(f'배열의 데이터 타입: {arr.dtype}')
print(f'배열의 차원 수: {arr.ndim}')
print(f'배열의 모양: {arr.shape}')
```

```python
# 10.
arr = np.random.randint(0, 100, (10,))
print(arr)
```

```python
## ndarray indexing과 slicing
# 1.
arr = np.array([10, 20, 30, 40, 50])
print(arr[2])
```

```python
# 2.
arr = np.arange(1, 10).reshape(3, 3).copy()
print(arr[1,1])
```

```python
# 3.
arr = np.array([10, 20, 30, 40, 50])
print(arr[1:4])
```

```python
# 4.
arr = np.arange(1, 13).reshape(3, 4).copy()
print(arr[0:2,1:3])
```

```python
# 5.
arr = np.arange(1, 19).reshape(3, 2, 3).copy()
print(arr[1,0,:])
```

```python
# 6.
arr = np.arange(1, 17).reshape(4, 4).copy()
print(np.diag(arr))
```

```python
# 7.
arr = np.arange(1, 11)
print(arr[(arr % 2) == 0])
```

```python
# 8.
arr = np.arange(1, 17).reshape(4, 4).copy()
print(arr[arr >= 5])
```

```python
# 9.
arr = np.arange(10)
arr[3:6] = -1
print(arr)
```

```python
# 10.
arr = np.arange(1, 17).reshape(4, 4).copy()
temp = np.zeros((4, 4))
temp[1:3,1:3] = arr[1:3,1:3]
print(temp)
```

```python
# 11.
arr = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60])
arr = arr[arr > 20]
arr = arr[arr < 50]
print(arr)
```

```python
# 12.
arr_2d = np.arange(1, 17).reshape(4, 4).copy()
print(arr_2d[np.ix_([0, 2], [1, 3])])
```

```python
# 13.
arr = np.array([[12, 15, 22, 29],
                [5, 18, 21, 30],
                [14, 17, 13, 20]])
arr[arr > 20] = 0
print(arr)
```

```python
# 14.
arr = np.array([[10, 15, 20, 25],
                [30, 35, 40, 45],
                [50, 55, 60, 65],
                [70, 75, 80, 85]])
print(np.where((arr > 40) & ((arr%2) == 0)))
arr1 = np.where((arr > 40) & ((arr%2) == 0), 0, arr)
print(arr1)
```
