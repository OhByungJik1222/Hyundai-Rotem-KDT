```python
import numpy as np
import pandas as pd

df_rating = pd.read_csv('/content/drive/MyDrive/KDT/data/ratings.csv')
df_movie = pd.read_csv('/content/drive/MyDrive/KDT/data/movies (1).csv')

display(df_rating.head(), df_movie.head())
```

```python
df_rating.info()
print()
df_movie.info()
```

```python
# 1.
print(df_rating['rating'].mean())
```

```python
df = pd.merge(df_rating, df_movie, on='movieId')
display(df)
df.info()
```

```python
# 2.
df_title = pd.DataFrame(df['title'].value_counts())
print(df_title.loc[df_title['count'] == df['title'].value_counts().max()])
```

```python
genres_list = list(df['genres'].apply(lambda x : x.split("|")))
df.drop('genres', axis=1)
df['genres'] = genres_list
```

```python
new_df = df.explode('genres')
display(new_df.head(10))
```

```python
# 3.
df_genre = new_df.groupby('genres')
print(df_genre.count().idxmax().max())
```

```python
# 4.
df_genre = new_df.groupby('genres')
print(df_genre.rating.mean().idxmax())
```

```python
new_df['year'] = pd.to_datetime(df['timestamp'], unit='s').dt.year
```

```python
# 5.
df_2015 = new_df.loc[new_df['year'] == 2015]
df_2015 =  df_2015.loc[df_2015['genres'] == 'Romance']
print(df_2015['rating'].mean())
```

```python
genres = list(df_movie['genres'].apply(lambda x : x.split("|")))

genres_len = []
for item in genres:
    genres_len.append(len(item))

df_movie['genres_len'] = genres_len
```

```python
# 6.
print(df_movie.loc[df_movie['genres_len'] == df_movie['genres_len'].max()]['title'])
```

```python
# 7.
df_title = pd.DataFrame(df['title'].value_counts())
print(df_title.loc[df_title['count'] == df['title'].value_counts().max()])
```

```python
# 8.
user_genres = new_df[['userId', 'genres']]

user_genres = user_genres.drop_duplicates()

genre_counts = user_genres.groupby('userId')['genres'].nunique()
df_userId = new_df.groupby('userId')
df_userId_max = df_userId.get_group(genre_counts.idxmax())

user_ratings = df_userId_max[['userId', 'genres', 'rating']]
genre_mean_ratings = user_ratings.groupby('genres')['rating'].mean()

print(genre_counts.idxmax(), genre_mean_ratings.idxmax())
```

```python
# 9.
df_10 = pd.DataFrame(df['title'].value_counts())
new_df = pd.merge(df, df_10, on='title')

group_by_count = new_df.loc[new_df['count'] >= 10].groupby('title')
print(group_by_count.rating.var().idxmin())
```
