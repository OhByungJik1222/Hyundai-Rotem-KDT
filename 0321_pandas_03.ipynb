```python
import numpy as np
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/KDT/data/auto-mpg.csv')
display(df)
print()

df = pd.read_csv('/content/drive/MyDrive/KDT/data/auto-mpg.csv', header=None)
df.columns = ['mpg', 'cylinder', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'name']
display(df)
print()

# 1. head() : DataFrame의 상위 5개의 row를 가져와서 DataFrame으로 생성
#             인자로 숫자를 사용할 수 있어요!
display(df.head())
print()

# 2. tail() : DataFrame의 하위 5개의 row를 가져와서 DataFrame으로 생성
#             인자로 숫자를 사용할 수 있어요!
display(df.tail())
print()

# 3. shape
print(df.shape)
print()

# 4. info() : DataFrame의 기본 정보를 화면에 출력할 수 있어요!
#             print()를 사용하지 않고 함수 자체로 출력기능을 가지고 있어요!
df.info()
print()

# 5. count() : DataFrame이 가지고 있는 각 열의 데이터의 개수를 Series로 리턴
#              count()는 유효한 값만 counting해요! NaN은 counting하지 않아요!
print(df.count())
print()

# 6. value_count() : Serie가 가지고 있는 함수에요!
#                    unique value가 몇개 있는지 알려줘요!
#                    NaN를 포함할건지 제외할건지를 옵션으로 명시 => dropna=True
print(df['origin'].value_counts())
print()

# 7. unique() : Series가 가지고 있는 함수! 중복을 제거하는 함수!
print(df['model_year'].unique())
print()

# 8. isin() : 어떤 값이 내가 정한 값들 안에 들어 있으면 True or False
#             Boolean indexing 할 때 많이 사용!
print(df['origin'].isin([1, 2]))
print()

# 제조국이 미국, EU가 아닌 차량의 mpg와 origin, name을 조회하세요!
display(df.loc[~df['origin'].isin([1, 2]), ['mpg', 'origin', 'name']])
```

```python
# 9. sort() : 정렬하는 함수
np.random.seed(1)

arr = np.random.randint(0, 10, (6, 4))
df = pd.DataFrame(arr)
df.columns = ['A', 'B', 'C', 'D']
df.index = pd.date_range('20250101', periods=6)
display(df)
print()

# 적절하게 DataFrame을 섞어볼거에요!
# 값을 섞을 때 사용하는 함수는 대표적으로 shuffle()
# index 는 mutable operation을 지원하지 않아요!
# 근데 shuffle은 mutable operation을 하는 함수에요!
# 고로 shuffle은 index를 섞는데 사용할 수 없어요!
# random_date = np.random.shuffle(df.index) Error
random_date = np.random.permutation(df.index)
print(random_date)
print()

# df.index = random_date 이렇게 하면 data는 안바뀌고 index만 바뀜 이렇게 하면 X
new_df = df.reindex(index=random_date, columns=['B', 'A', 'D', 'C'])
display(new_df)
print()

# 이제 정렬을 해보아요!
# index를 이용한 정렬과 value를 이용한 정렬이 있어요!
# index를 이용한 정렬부터 알아보아요!
# 당연히 정렬할 때 axis를 설정
display(new_df.sort_index(axis=1, ascending=True))
print()

display(new_df.sort_index(axis=0, ascending=False))
print()

# value를 이용한 정렬
display(new_df.sort_values(by='B'))
```

```python
# DataFrame의 기술 통계(Descriptive Statistics)를 위한 함수들이 있어요!
# sum(). mean(), max(), min(), count(), std(), ...
# NumPy 때는 axis를 사용하지 않으면 => 전체 ndarray를 대상으로 함수를 적용
# Pandas는 axis를 사용하지 않으면 => 기본 axis=0 (row 방향)을 기준
data = np.array([[2, np.nan], [7, -3], [np.nan, np.nan], [1, -2]])
df = pd.DataFrame(data, columns=['one', 'two'], index=['A', 'B', 'C', 'D'])
display(df)
print()

print(df.sum()) # df.sum(axis=0, skipna=True) -> Series로 리턴
print()

print(df.sum(skipna=False)) # NaN은 값이 없기 때문에 (값을 알 수 없기 때문에) sum하면 그 값도 알 수 없다.
print()

print(df.sum(axis=1))
print()

# 특정 행, 열에 대한 합
print(df['two'].sum()) # 열
print(df.loc['B'].sum()) # 행
print()

# 평균을 구할 때는 mean(), NaN은 연산에서 아예 포함 X
print(df.mean())
print()

# NaN 결측값 처리
df['one'] = df['one'].fillna(value=df['one'].mean())
display(df)
```

```python
# 데이터 처리를 위한 함수들을 알아봏기 위해 데이터 셋을 사용할거에요!
# MPG Date Set(자동차 연비)과 Titanic Data Set을 사용할 거에요!
# Titanic Data Set은 유명한 Data Set으로
# 다양한 경로로 데이터를 구해서 사용할 수 있어요!
# 그 중 우리는 seaborn library에서 제공하는 데이터 셋을 사용할거에요!

# 1. 결치값 확인!
import numpy as np
import pandas as pd
import seaborn as sns

# raw data set을 가져와요!
df = sns.load_dataset('titanic')

display(df.head())
df.info()
print()

# 각 컬럼의 결측치 개수를 알고 싶어요! => 많이 쓰는 방법은 아니에요!
print(df['deck'].value_counts(dropna=False))
print()


# 결측치를 직접 찾는 방법으로 가장 많이 사용하는 함수가 있어요!
display(df.isnull()) # notnull()도 있어요!
print()
display(df.isnull().sum(axis=0))
print()

# 2. 결측치 삭제 (결측치를 행 단위와 열 단위로 삭제하는법을 알아보았어요!)
# 결측치 확인이 끝나면 특정 조건을 이용해서 결측치가 포함된 column을 삭제할 수 있어요!
# deck의 경우 900개 중 700개 가량이 NaN이기 때문에 임의의 값으로 채우는것 보단 삭제하는 것이 났다.
new_df = df.dropna(axis=1, inplace=False) # 이렇게 하면 NaN이 포함된 모든 column이 삭제된다.
display(new_df.head())
print()
display(new_df.isnull().sum(axis=0))
print()

new_df = df.dropna(axis=1, inplace=False, thresh=500) # thresh : NaN의 개수가 500이 넘으면 삭제한다.
display(new_df.head())
print()
display(new_df.isnull().sum(axis=0))
print()

# 가장 문제가 되는 컬럼을 삭제 처리했어요!
# NaN을 포함하는 행을 찾아서 삭제하거나 대체해야 해요!
df2 = new_df.dropna(axis=0, how='any') # 행 안에 NaN이 하나라도 있으면 삭제!
display(df2.isnull().sum(axis=0))
print()

df2 = new_df.dropna(subset=['age'], axis=0, how='any') # subset을 사용해서 제한도 가능!
display(df2.isnull().sum(axis=0))

```
